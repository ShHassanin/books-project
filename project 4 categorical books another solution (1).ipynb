```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url="https://books.toscrape.com/"
response= requests.get(url)
```

```python
soup= BeautifulSoup(response.text,"html.parser")
```

```python
links= soup.find('ul', class_="nav nav-list").find_all('a')
links
```

```python
links.remove(links[0])
links
```

```python
with open('Books.csv','w',newline='', encoding='utf-8') as f:
    writer= csv.DictWriter(f, fieldnames=['categoryID','category','bookName','price','rating'])
    writer.writeheader()
    
    categoryID =0
    for link in links:
        url2=url+link.get('href')
        category= link.get_text().strip()
        categoryID +=1
        response2= requests.get(url2)
        soup2= BeautifulSoup(response2.text,"html.parser")
        articles= soup2.find_all('article', attrs={'class':"product_pod"})
        nextt= soup2.find('li', attrs={'class':"next"})
        
        while nextt != None:

            url3=url2.replace('index.html', nextt.find('a').get('href'))
            print(nextt, url3)
            response3= requests.get(url3)
            soup3= BeautifulSoup(response3.text,"html.parser")
            articles2=soup3.find_all('article', attrs={'class':"product_pod"})
            for r in articles2:
                articles.append(r)
            print(category,url3)
            nextt= soup3.find('li', attrs={'class':"next"})

            
            
        for article in articles:
                price= float((article.find('p', attrs={'class':'price_color'}).get_text())[2:])
                bookName= article.find('h3').find('a').get('title')
                stars = article.find('p').get('class')[1]
                starsdic= {'One':1,'Two':2,'Three':3,"Four":4,'Five':5}
                rating= int(starsdic[stars])
                print(bookName,price,rating, category )
                
                writer.writerow({
                    'categoryID':categoryID,
                    'category':category,
                    'bookName':bookName,
                    'price':price,
                    'rating':rating
                })
```
